# El Perceptr贸n
<a id="ElPerceptron"></a>

Dado un punto  $\mathbf{x} \in R^d$ , el *modelo de perceptr贸n* le asigna la clase +1 o -1 de acuerdo a la siguiente expresi贸n

$$
h(\mathbf{x}) = signo \left( \sum_{i=1}^d w_i x_i + b \right). 
$$

A los efectos de hacer la notaci贸n m谩s compacta, es habitual escribir dicha expresi贸n en *coordenadas homog茅neas*. Llamando $w_0$ a $b$ y agregando una coordenada con valor constante 1 a  $\mathbf{x}$, es decir $\mathbf{x} = [1, x_1, \ldots , x_d] $ la expresi贸n anterior queda

$$
h(\mathbf{x}) = signo \left( \sum_{i=0}^d w_i x_i \right) = signo \left(\mathbf{w}^T \mathbf{x} \right).
$$  

Para encontrar los par谩metros del modelo es necesario contar con un conjunto de puntos de entrenamiento $\{\mathbf{x}_1, \ldots, \mathbf{x}_N\}$ con sus correspondientes etiquetas $\{y_1, \ldots, y_N\}$. El *algoritmo del perceptr贸n* recorre en forma iterativa el conjunto de puntos de entrenamiento y en cada paso $t$ se fija si el modelo clasifica correctamente al punto $\mathbf{x}(t)$. En caso afirmativo no hace nada y en caso negativo actualiza los pesos de acuerdo a la siguiente regla:

$$
\mathbf{w}(t+1)=\mathbf{w}(t)+y(t)\mathbf{x}(t)
$$

siendo $y(t)$ la etiqueta correspondiente al punto $\mathbf{x}(t).$

## Ejercicio 1
<a id="Ejercicio1"></a>

Cada vez que se actualizan los pesos del perceptr贸n 

$$
\mathbf{w}(t+1)=\mathbf{w}(t)+y(t)\mathbf{x}(t)
$$

la frontera de decisi贸n determinada por los pesos se mueve a una direcci贸n tendiente a clasificar la muestra actual correctamente. Asumiendo que la muestra $\mathbf{x}(t)$ es mal clasificada en el paso actual      

**a)** Mostrar que $y(t) \mathbf{w}^T(t) \mathbf{x}(t)<0.$

**RESPUESTA:**  

Como la muestra x(t) es mal clasificada significa que y(t) tiene signo contrario a $\mathbf{w}^T(t) \mathbf{x}(t)$, por lo tanto su multiplicaci贸n ser谩 negativa y menor a 0. 

**b)** Mostrar que $y(t)\mathbf{w}^T(t+1)\mathbf{x}(t) > y(t)\mathbf{w}^T(t)\mathbf{x}(t).$

**RESPUESTA:**
Sustituyendo $\mathbf{w}(t+1)=\mathbf{w}(t)+y(t)\mathbf{x}(t)$ tenemos:

$$y(t)(\mathbf{w}(t)+y(t)\mathbf{x}(t))^T\mathbf{x}(t) > y(t)\mathbf{w}^T(t)\mathbf{x}(t)$$

Usando las propiedades de Transitividad:

$$y(t)\mathbf{w}(t)^T\mathbf{x}(t) + y(t)\mathbf{x}(t)^Ty(t)^T\mathbf{x}(t) > y(t)\mathbf{w}^T(t)\mathbf{x}(t)$$

Como de ambos lados tenemos $y(t)\mathbf{w}(t)^T\mathbf{x}(t)$ se traduce en:

$$y(t)\mathbf{x}(t)^Ty(t)^T\mathbf{x}(t) >0$$

Como $y(t)$ es un escalar podemos moverlo de lugar:

$$y(t)y(t)^T\mathbf{x}(t)^T\mathbf{x}(t) >0$$

Escribi茅ndolos como su norma vemos notoriamente que se cumple:

$$|y(t)| ^2 \|x(t)\| ^2 >0$$


**c)** Argumentar que la actualizaci贸n de $\mathbf{w}(t)$ a $\mathbf{w}(t+1)$ es un movimiento en la direcci贸n correcta en cuanto a clasificar a $\mathbf{x}(t)$ correctamente se refiere.

**RESPUESTA:**   

Al cumplirse ambos puntos anteriormente demostrados se concluye que el signo de y(t) pasa de ser distinto al de 梆()() a ser igual, por lo tanto ahora x(t) estar谩 bien clasificada.

**a)** Mostrar que las regiones del plano donde $h(\mathbf{x})= +1$ y $h(\mathbf{x})= -1$ est谩n separadas por una recta. Si se expresa dicha recta mediante la ecuaci贸n $x_2 = mx_1 + n$, escribir la pendiente $m$ y la intersecci贸n $n$ en funci贸n de $w_0$, $w_1$, $w_2$    

**RESPUESTA:**  

La recta que separa el signo de $\mathbf{w}^T\mathbf{x}$ de ser positivo o negativo es la que surge de imponer $\mathbf{w}^T\mathbf{x}=0$. 
En este caso tenemos:
$$\mathbf{w}^T\mathbf{x}=w_0+w_1x_1+w_2x_2=0$$
Separando $x_2$:
$$x_2=-\frac{w_0}{w_2}-\frac{w_1}{w_2}x_1$$
Siendo $n=-\frac{w_0}{w_2}$ y $m=-\frac{w_1}{w_2}$ tenemos:
$$x_2 = mx_1 + n$$

**b)** Graficar la frontera de decisi贸n para los casos $\mathbf{w}=[1,2,3]^T$ y $\mathbf{w}=-[1,2,3]^T$ 

En dimensiones mayores a dos, las regiones +1 y -1 est谩n separadas por un hiperplano, la generalizaci贸n de una recta.

### Generar conjunto de entrenamiento 

Completar el c贸digo de la funci贸n `generar_conjunto_entrenamiento()`. Dentro de la misma se deber谩 eligir aleatoriamente una recta del plano como funci贸n objetivo $f$. La recta elegida divide al plano en dos regiones, a los puntos pertenecientes a una de ellas se les asignar谩 la etiqueta +1 y a los pertenecientes a la otra se le asignar谩 -1. Se asumir谩 que el espacio $\mathcal{X} = [-1,1] \times [-1,1]$. Los puntos $\mathbf{x}_n$ del conjunto de entrenamiento ser谩n elegidos de manera aleatoria y sus etiquetas $y_n$ ser谩n asignadas evaluando la funci贸n objetivo para cada $\mathbf{x}_n$. De esta forma se generar谩 un conjunto de entrenamiento de tama帽o $N=20$. 

```python
def generar_conjunto_entrenamiento(cantidad_muestras, d=2, semilla=None):
    """ Esta funci贸n genera un conjunto de entrenamiento de tama帽o cantidad_muestras
    Entrada:
        cantidad_muestras: cantidad de muestras que se devuelven
        d: dimensi贸n en la que viven las muestras generadas
        semilla: valor que se le asigna al m茅todo random.seed()
    Salida:
        w_rand: vector de tama帽o d+1 que contiene los par谩metros 
                de la funci贸n objetivo elegida aleatoriamente
        X: matriz de tama帽o (cantidad_muestras x (d+1)) que contiene las 
           muestras de entrenamiento generadas
        y: vector de tama帽o cantidad_muestras que contine las etiquetas de las 
           muestras generadas
    """
    def sign_function(x):
        if x > 0:
            return 1
        elif x == 0:
            return 0
        else:
            return -1
    
    # El espacio se restringe a la regi贸n [-1, 1] x [-1, 1]
    x_min = -1
    x_max = 1

    if semilla is not None:
        np.random.seed(semilla)

    #######################################################
    ######## EMPIEZA ESPACIO PARA COMPLETAR CODIGO ########
    #######################################################

    # Se elige aleatoriamente una recta del plano como funci贸n objetivo f
    
    w_rand = np.random.rand(d+1)

    # Se elige aleatoriamente N puntos pertenecientes a la regi贸n de inter茅s. 
    # Se agregan los unos correspondientes a las coordenadas homog茅neas.
    
    X = (np.random.rand(cantidad_muestras,d+1)*(x_max-x_min)) - np.ones((cantidad_muestras,d+1))
    X[:,0]=np.ones((1,cantidad_muestras))
    # Se generan las salidas utilizando el modelo del perceptr贸n.
    y=w_rand@X.T
    y = np.sign(y)
    #######################################################
    ######## TERMINA ESPACIO PARA COMPLETAR CODIGO ########
    #######################################################

    return w_rand, X, y


N = 20
w, X, y = generar_conjunto_entrenamiento(N, semilla=42)

validar_resultado('p01e03', w, X, y)
```

### Implementar el perceptr贸n
<a id="implementarPerceptron"></a>

Implementar el m茅todo `entrenar_perceptron()`.

```python
def entrenar_perceptron(X, y, w_inicial=None, max_iter=500):
    """
    Entrada:
        X: matriz de (Nxd+1) que contiene las muestras de entrenamiento
        y: etiquetas asociadas a las muestras de entrenamiento
        max_iter: m谩xima cantidad de iteraciones que el algoritmo 
                  puede estar iterando
        w_inicial: inicializaci贸n de los pesos del perceptr贸n
        
    Salida:
        w: par谩metros del modelos perceptr贸n   
        error: lista que contiene los errores cometidos en cada iteraci贸n
    """

    if w_inicial is None:
        # Se inicializan los pesos del perceptr贸n
        w = np.random.rand(X.shape[1])  # w = np.zeros(d+1)
    else:
        w = w_inicial

    #######################################################
    ######## EMPIEZA ESPACIO PARA COMPLETAR CODIGO ########
    #######################################################
    error=[]
    i=0
    while (i != max_iter) :
        error.append(np.count_nonzero(y-np.sign(w@X.T)))
        if error[-1]!=0 :
            ind=(y!=np.sign(w@X.T))
            errores_y=y[ind]
            errores_X=X[ind,:]
            w=w+errores_y[0]*errores_X[0,:]
            i+=1
        else :
            i=max_iter
    
    #######################################################
    ######## TERMINA ESPACIO PARA COMPLETAR CODIGO ########
    #######################################################

    return w, error
```

```python
w_perceptron, error = entrenar_perceptron(X, y)
print('El perceptr贸n finaliz贸 en la iteraci贸n %d' % len(error))

# El error de entrenamiento que con el w encontrado deber铆a ser cero.
print('El error de entrenamiento es %f' % error[-1])
```

c) Usando b) graficar una estimaci贸n de $P(\vert \nu - \mu \vert > \epsilon )$ en funci贸n de $\epsilon$ junto con la cota de Hoeffding (en la misma figura)

```python
# Cota de Hoeffding
epsilons = np.arange(0., 1.05, 0.1)
cota_Hoeffding = 2.0 * np.exp(-2 * (epsilons**2) * N)
mu = 0.5


def calcular_probabilidad_empirica(V_j, mu, epsilons):
    '''
    Calcula la probabilidad P(|nu - mu| > epsilon ) para un experimento
    Entrada:
        V_j: vector de tama帽o N que almacena los resultados de N experimentos para una de las monedas
        mu: valor de mu
        epsilons: valores de epsilon para los que se desea calcular la probabilidad
    Salida: 
        probabilidad_empirica: vector de tama帽o igual al largo de epsilons que contiene las probabilidades
                             calculadas
    '''
    probabilidad_empirica = np.zeros(len(epsilons))
    for i, eps in enumerate(epsilons):

        #######################################################
        ######## EMPIEZA ESPACIO PARA COMPLETAR CODIGO ########
        #######################################################

        # Se suman los experimentos que distaron de mu m谩s de epsilon por debajo
        TF_1=((mu-V_j)>eps)
        sum_1=np.count_nonzero(TF_1 == True)
        
        # Se suman los experimentos que distaron de mu m谩s de epsilon por arriba
        TF_2=((V_j-mu)>eps)
        sum_2=np.count_nonzero(TF_2 == True)
        # Se calcula la probabilidad para el epsilon correspondiente
        probabilidad_empirica[i]=((sum_1+sum_2)/len(V_j))

        #######################################################
        ######## TERMINA ESPACIO PARA COMPLETAR CODIGO ########
        #######################################################

    return probabilidad_empirica


resultado_empirico_moneda0 = calcular_probabilidad_empirica(
    V[:, 0], mu, epsilons)
resultado_empirico_moneda_aleatoria = calcular_probabilidad_empirica(
    V[:, 1], mu, epsilons)
resultado_empirico_moneda_menos_veces = calcular_probabilidad_empirica(
    V[:, 2], mu, epsilons)

plt.figure(figsize=(30, 10))
plt.subplot(1, 3, 1)
plt.plot(epsilons, cota_Hoeffding, drawstyle='steps', label='Cota Hoeffding')
plt.plot(epsilons,
         resultado_empirico_moneda0,
         drawstyle='steps',
         label='Resultado empirico')
plt.grid()
plt.xlabel('epsilon', fontsize=20)
plt.title('Cota de Hoeffding', fontsize=20)
plt.legend(fontsize=20)
plt.subplot(1, 3, 2)
plt.plot(epsilons, cota_Hoeffding, drawstyle='steps', label='Cota Hoeffding')
plt.plot(epsilons,
         resultado_empirico_moneda_aleatoria,
         drawstyle='steps',
         label='Resultado empirico')
plt.grid()
plt.xlabel('epsilon', fontsize=20)
plt.title('Cota de Hoeffding', fontsize=20)
plt.legend(fontsize=20)
plt.subplot(1, 3, 3)
plt.plot(epsilons, cota_Hoeffding, drawstyle='steps', label='Cota Hoeffding')
plt.plot(epsilons,
         resultado_empirico_moneda_menos_veces,
         drawstyle='steps',
         label='Resultado empirico')
plt.grid()
plt.xlabel('epsilon', fontsize=20)
plt.title('Cota de Hoeffding', fontsize=20)
plt.legend(fontsize=20)
plt.show()
```
